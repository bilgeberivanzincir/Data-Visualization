```python
import seaborn as sns
flights = sns.load_dataset("flights")
df=flights.copy()
df.head()
```

```python
df.shape
```

```python
df.passengers.describe()
```

```python
#pivot table oluşturma, heatmap bunu kullanır
df.pivot_table("passengers","month","year")
```

```python
sns.heatmap(df.pivot_table("passengers","month","year"));
```

```python
#headmap haritasını çiz
#yılın hangi ayında kaç yolcu taşındığını gösterir
#annot=True, sayıları gösterir
#fmt="d" sayıları virgülden sonra 0 basamaklı gösterir
#linewidths=1, çizgiler arasındaki boşluk
sns.heatmap(df.pivot_table("passengers","month","year"),annot=True,fmt="d",linewidths=1);
```
