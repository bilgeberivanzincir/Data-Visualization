```python
import seaborn as sns
tips = sns.load_dataset('tips')
df=tips.copy()
df.head()
```

#iki sayısal değişkenin birlikte oldukları yapıların değerlendirilmesine korelasyon denir. 

```python
#dağılım grafiği
sns.scatterplot(x="total_bill", y="tip", data=df);
```

```python
#doğrusal ilişkinin gösterilmesi
#lm=lineer model
import matplotlib.pyplot as plt
sns.lmplot(x="total_bill", y="tip", data=df);
```

```python
#çaprazlama ekleme
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=df);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=df);
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", row="sex", data=df);
```

```python
#scatterplot matrisi
import seaborn as sns
iris = sns.load_dataset('iris')
df=iris.copy()
df.head()
```

```python
df.dtypes
```

```python
df.shape
```

```python
#veri setindeki değişkenlerin birbirleriyle olan ilişkilerini görmek için kullanılır
#scatterplot matrisi pairplot ile gösterilir
sns.pairplot(df);
```

1. iki değişken birbirinde farklı köşelerde toplanıyorsa, çaprazlama durumuyla gözlemlenebilen alt gruplar vardır. 

2. eğer oluşan grafik toz bulutu şeklindeyse iki değişken arasında ilişki yoktur demektir.

```python
sns.pairplot(df, hue="species");
```

```python
sns.pairplot(df, hue="species", markers=["o", "s", "D"]);
```

```python
#grafiğe doğru ekleme
sns.pairplot(df, kind="reg", hue="species");
```
