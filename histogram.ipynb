```python
import seaborn as sns
diamonds= sns.load_dataset('diamonds')
df=diamonds.copy()
df.head()
```

```python
#yoğunluk false
sns.distplot(df.price, kde= True);
```

```python
#bins aralıklarla ilgili (10 çubuk olacak demek)
sns.distplot(df.price,bins=10,kde=False)
```

```python
#histogram ile yoğunluk birlikte gösteriliyor
sns.distplot(df.price,kde=True)
```

```python
sns.kdeplot(df.price, shade=True)
```

# Çaprazlamalar

```python
#nasıl olmuş
(sns
 .FacetGrid(df,
               hue="cut",
               height=5,
               xlim=(0, 10000))
.map(sns.kdeplot,"price", fill=True)
.add_legend()
);
```

```python
sns.catplot(x="cut",y="price", hue="color", kind="point", data=df);
```
